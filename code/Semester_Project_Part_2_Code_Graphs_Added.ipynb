```python
dict_traveler = {'Friends': 1, 'Business': 2, 'Families': 3, 'Couples': 4, 'Solo': 5}
```

```python
df['Num Traveler type'] = df['Traveler type'].map(dict_traveler)
```

```python
df
```

```python
df['User country'].unique()
```

```python
dict_UC = {}
i = 1
for temp in df['User country'].unique():
    dict_UC[temp] = i
    i = i+1
```

```python
dict_UC
```

```python
df['Num User country'] = df['User country'].map(dict_UC)
```

```python
dict_hotelname = {}
i = 1
for temp in df['Hotel name'].unique():
    dict_hotelname[temp] = i
    i = i+1
df['Num Hotel name'] = df['Hotel name'].map(dict_hotelname)
```

```python
df
```

```python
dict_pdofstay
df['Num Period of stay'] = df['Period of stay'].map(dict_pdofstay)
#df.iloc[:,-5:]
df
```

```python
df['Dummy Pool'] = pd.get_dummies(df['Pool'], prefix='Pool' ,drop_first=True)
df['Dummy Gym'] = pd.get_dummies(df['Gym'], prefix='Gym' ,drop_first=True)
df['Dummy Tennis'] = pd.get_dummies(df['Tennis court'], prefix='Tennis' ,drop_first=True)
df['Dummy Spa'] = pd.get_dummies(df['Spa'], prefix='Spa' ,drop_first=True)
df['Dummy Casino'] = pd.get_dummies(df['Casino'], prefix='Casino' ,drop_first=True)
df['Dummy Free Internet'] = pd.get_dummies(df['Free internet'], prefix='Free internet' ,drop_first=True)
```

```python
dict_reviewmonth = {}
i = 1
for temp in df['Review month'].unique():
    dict_reviewmonth[temp] = i
    i = i+1
df['Num Review month'] = df['Review month'].map(dict_reviewmonth)
df.iloc[:,5:]
```

```python
dict_usercontinent = {}
i = 1
for temp in df['User continent'].unique():
    dict_usercontinent[temp] = i
    i = i+1
df['Num User continent'] = df['User continent'].map(dict_usercontinent)
```
